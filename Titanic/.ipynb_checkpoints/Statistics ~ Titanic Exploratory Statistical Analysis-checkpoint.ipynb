### Merge Data

```python
train = pd.read_csv("titanic_training.csv", header = 0) 
data = pd.DataFrame.copy(train)
# test = pd.read_csv("titanic_test.csv", header = 0)
# ID = test['PassengerId'] # Extract ID Names From Test Set
# data = pd.concat([train, test], axis = 0)
data.shape
```

# Preprocess

# Ticket

```python
data["Ticket"] = data["Ticket"].apply(len)
data["LenTicket"] = data["Ticket"].ffill().bfill()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(np.array(data.Ticket))
data["Ticket"] = le.transform(data.Ticket)
```

# Cabin

```python
for x in data:
    data["Cabin"] = data["Cabin"].str[0]
```

```python
def transform(char): 
    if char == "0":
        return np.nan 
    if char == "A":
        return 1 
    if char == "B":
        return 2 
    if char == "C":
        return 3 
    if char == "D":
        return 4 
    if char == "E":
        return 5 
    if char == "F":
        return 6 
    if char == "G":
        return 7 
    if char == "T":
        return np.nan

data['Cabin'] = data['Cabin'].apply(transform)
data["Cabin"].fillna(data.groupby("Fare")["Cabin"].transform("mean"), inplace=True)
```

```python
data['Title'] = 'Other'
data.loc[['Master.' in title for title in train.Name], 'Title'] = 'Master'
data.loc[['Miss.' in title for title in train.Name], 'Title'] = 'Miss'
data.loc[['Mr.' in title for title in train.Name], 'Title'] = 'Mr'
data.loc[['Mrs.' in title for title in train.Name], 'Title'] = 'Mrs'
data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(np.array(data.Title))
data["Title"] = le.transform(data.Title)
data.head()
```

### Delete Columns

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(np.array(data.Embarked))
data["Embarked"] = le.transform(data.Embarked)
```

```python
data['Sex'] = data.Sex.map({'male':1, 'female':0})
```

```python
del data['Name'], data['Ticket'], data['PassengerId']
```

```python
# Fill Missing Fare With Median Fare Of Each Pclass
data["Fare"].fillna(data.groupby("Pclass")["Fare"].transform("median"), inplace=True)

# Fill Missing Age With Median Age Based On The Median Of A Person's Title
data["Age"].fillna(data.groupby("Title")["Age"].transform("median"), inplace=True)
```

```python
le.fit(np.array(data.Title))
data["Title"] = le.transform(data.Title)
```

```python
data = data.fillna(train.median(), inplace = True)  # Impute Missing Data
data = data.astype(float) # Convert DF Type to Float

"""Split Data"""
train = data[0:len(train)]
# test = data[len(train):]

"""Normalize Data for Faster Computation"""
# train = train/train.max().astype(np.float64)
# test = test/test.max().astype(np.float64)
```

```python
def bar_chart(feature):
    survived = train[train['Survived']==1][feature].value_counts()
    dead = train[train['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))    
```
